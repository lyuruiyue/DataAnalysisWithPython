# Question 2

```python
with open('ques2.txt', 'r') as f:
    words_count = 0
    word_length = []
    length = 0
    for line in f.readlines():
        words = line.split(' ')
        words_count = len(words)
        for word in words:
            length = len(word)
            word_length.append(length)
print ('The number of words is:', words_count)
sum=0
for l in word_length:
    sum = sum+l
ave = sum/words_count
print ('The average is:',ave)
word_length.sort()
median = 0
if words_count %2 ==1:
    print ('The median is:',word_length[int((words_count-1)/2)])
if words_count %2 ==0:
    print ('The median is:',word_length[int(((words_count/2)-1)+int((words_count/2)))/2])
count_appear = dict((a, word_length.count(a)) for a in word_length)
for k,v in count_appear.items():
    if v == max(count_appear.values()):
        print('The mode is', k)
f.close
```

```python
import os
root_dir = 'jsons'
dic = {}
for parent, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        filepath = parent + os.sep + filename
        count1 = 0
        count3 = 0
        if filepath.endswith(".json"):
            myfile = open(filepath)
            data = json.load(myfile)
            segments = data.get('retargetingSegments')
            if type(segments) == list:
                for subDic in segments:
                    value = subDic.get('segmentUid')
                    if value[0] == '1':
                        count1 += 1
                    else:
                        count3 += 1
        subCount = {'first': count1, ' third': count3}
        dic.update({filename[:-5]:subCount})
    print(dic)
    
            
    
```
